```python
import numpy as np
from scipy.signal import firwin, freqz
import matplotlib.pyplot as plt

# Define the filter parameters
num_taps = 8  # Number of taps
cutoff_freq = 2000  # Cutoff frequency in Hz
nyquist_rate = 10000  # Nyquist rate (half of the sampling frequency)

# Calculate the normalized cutoff frequency
cutoff_norm = cutoff_freq / nyquist_rate

# Design the filter using firwin
filter_coeffs = firwin(num_taps, cutoff_norm)

# Plot the frequency response of the filter
w, h = freqz(filter_coeffs)
freq = w * nyquist_rate / (2 * np.pi)  # Convert angular frequency to Hz

plt.figure()
plt.plot(freq, 20 * np.log10(abs(h)), label='Filter response')
plt.axvline(cutoff_freq, color='red', label='Cutoff frequency')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Frequency Response of 8-tap Low-Pass Filter')
plt.legend()
plt.grid()
plt.show()

print('Filter Coefficients:')
print(filter_coeffs)
```
